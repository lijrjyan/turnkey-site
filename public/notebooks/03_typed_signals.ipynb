# Turnkey: typed signals
A Policy asks for a typed value; a provider materializes it once and Turnkey reuses equal requests.

```python
from dataclasses import dataclass
from turnkey.methods import MethodContext, Request

@dataclass(frozen=True)
class LengthSignal(Request[float]):
    prompt: str

class LengthProvider:
    request_type = LengthSignal
    model_forwards_per_call = 0
    def provide(self, request: LengthSignal) -> float:
        return len(request.prompt) / 100.0

with MethodContext([LengthProvider()]) as context:
    first = context.get(LengthSignal('Bring Your Own Detector'))
    second = context.get(LengthSignal('Bring Your Own Detector'))
    uses = context.uses

assert first == second
assert [use.cache_hit for use in uses] == [False, True]
[(use.cache_hit, use.model_forwards) for use in uses]
```
